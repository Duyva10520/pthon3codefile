```python
def build_vrf_config(vrf_name, rd, rt_import='', rt_export=''):
    config = f"vrf name {vrf_name}\n"
    config += f"  rd {rd}\n"
    if rt_import:
        config += f"  route-target import {rt_import}\n"
    if rt_export:
        config += f"  route-target export {rt_export}\n"
    config += "!\n"
    return config

config1 = build_vrf_config('techcombank', '11', rt_import='12')
print(config1)
config2 = build_vrf_config('techcombank', '11', rt_export='12')
print(config2)
```

```python
def build_vrf_config(vrf_name, rd, rt_import='', rt_export=''):
    if not rt_import:
        rt_import = rd
    if not rt_export:
        rt_export = rd
    
    return f"""
    vrf name {vrf_name}
    rd {rd}
    address-family ipv4
    route-target import {rt_import}
    route-target export {rt_export}
    """

config = build_vrf_config('techcombank', '11')
print(config)
```

```python
def build_vrf_config(vrf_name, rd, rt_import='', rt_export=''):
    rt_import = rd if not rt_import else rt_import
    rt_export = rd if not rt_export else rt_export
    
    return f"""
    vrf name {vrf_name}
    rd {rd}
    address-family ipv4
    route-target import {rt_import}
    route-target export {rt_export}
    """

config = build_vrf_config('techcombank', '11')
print(config)
```
